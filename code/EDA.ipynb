## Class imbalance analysis

```python
counts= df["Class"].value_counts()
plt.bar(counts.index,counts.values)
plt.title("Target distribution")
plt.show()
```

```python
counts = df["Class"].value_counts(normalize=True) * 100
plt.bar(counts.index, counts.values)
plt.ylabel("Percentage")
```

Target value distribution plots show <1% of class values of 1 Therefor:
1. Accuracy is misleading
2. Need precision, Recall, F1, ROC in **model evaluations**
3. Need SMOTE / class weighting in **preprocessing**

## Transaction amount analysis

```python
plt.subplot(2,1,1)
plt.hist(df["Amount"], bins=50)
plt.title("All transactions")

plt.subplot(2,1,2)
plt.hist( df[df["Class"]==0]["Amount"], bins=50, alpha=1, label="Normal")
plt.hist( df[df["Class"]==1]["Amount"], bins=50, alpha=1, label="Fraud")
plt.title("Fraud vs. normal")
plt.legend()
plt.yscale('log') 



plt.tight_layout()
plt.show()
```

```python
plt.subplot(2,1,1)
plt.hist( df[df["Class"]==1]["Amount"], bins=50)
plt.title("Fraud transactions")

plt.subplot(2,1,2)
plt.hist( df[df["Class"]==0]["Amount"], bins=50)
plt.title("Normal transactions")

plt.tight_layout();
plt.show()
```

Comparing the 4 plots above:

1. **Fraudsters try to blend in:** Normal and fraud transactions are both heavily skewed to the right.  
2. **Huge amounts are most likely normal:** Fraud transactions stop at ~2500 but normal transactions go up to 25k.  
3. **Class imbalance:** Normal transaction plot is very similar to all transaction plot, meaning very few transactions were fraud.

```python
print(df[df["Class"]==0]["Amount"].min())
print(df[df["Class"]==1]["Amount"].min())
```

**0$ fraud is also possible:** fraudulants might probably check if the stolen card is active and working. model has to rely on features other than 0$ purchase to tell if a transaction is fraud.
